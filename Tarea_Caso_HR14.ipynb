# Aplicar una regresión lineal para explorar la relación entre EmpSatisfaction (variable independiente) y PerfScoreID (variable dependiente).

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns

#variable independiente
x = df[["EmpSatisfaction"]]

# Variable dependiente
y = df["PerfScoreID"]

# Crear modelo
model = LinearRegression()
model.fit(x, y)

# Predicciones
y_pred = model.predict(x)

#Calcular R2
r2 = r2_score(y, y_pred)

print("Intercepto:", model.intercept_)
print("Coeficiente:", model.coef_[0])
print("R2", r2)
```

#  Representar gráficamente la relación y la línea de tendencia.

```python
from matplotlib.lines import lineStyles
plt.scatter(y, y_pred, label= "Predicciones")
plt.plot([y.min(), y.max()], [y.min(), y.max()], linestyle="--")
plt.title("Modelo de regresión leneal")
plt.xlabel("Predicciones")
plt.ylabel("Valor real")
```
